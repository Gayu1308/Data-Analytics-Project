```python
fig = px.histogram(df, x='total_cases', nbins=30, title='Distribution of Total Cases ')
fig.update_layout(
    xaxis_title='Total Cases',
    yaxis_title='Location',
    title_x=0.5,  # Center the title
    bargap=0.1,   # Add space between bars
    template='plotly_white' # Use a clean template
)
fig.show()
fig.write_html("distribution_of_total_cases.html")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='location', order=df['location'].value_counts().index)
plt.title('Leading Location')
plt.xlabel('location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
continent_cases = df.groupby('continent')['total_cases'].sum().reset_index()

fig = px.pie(
    continent_cases,
    values='total_cases',
    names='continent',
    title='Percentage of Total Cases by Continent'
)
fig.show()
```

```python
corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(12, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')  # Try 'viridis', 'magma', 'icefire', etc.
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```
